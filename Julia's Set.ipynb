```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def f(z, a, c):
    return a*z**2 + c

data = np.zeros((10000, 10000))
```

```python
points = []
a = np.array(range(-5000, 5000))/(5000*0.65)
for i in range(-5000, 5000):
    points.append(a+i/(5000*0.65)*1j)
    
points = np.array(points)
```

```python
for _ in range(20):
    points = f(points, 1, 0.1 -0.5j)
    data += np.ones((10000,10000))*((np.real(points)**2+np.imag(points)**2)<1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(data, cmap=cm.hot)
```
