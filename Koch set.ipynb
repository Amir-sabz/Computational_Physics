```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
rot_plus = 0.5*np.array([[1,-sqrt(3)],
                        [sqrt(3),1]])

rot_minus = 0.5*np.array([[1,sqrt(3)],
                        [-sqrt(3),1]])

Identity = np.array([[1,0],
                    [0,1]])

def Transform(arr,a,R,T):
    return a*np.matmul(R, arr)+np.array(T)
```

```python
def f(n):
    if n == 0:
        return np.array([[0,1],[0,0]])
    else:
        part1 = Transform(f(n-1), 1/3, Identity, [[0],[0]])
        part2 = Transform(f(n-1), 1/3, rot_plus, [[1/3],[0]])
        part3 = Transform(f(n-1), 1/3, rot_minus, [[1/2],[sqrt(3)/6]])
        part4 = Transform(f(n-1), 1/3, Identity, [[2/3],[0]])
        return np.hstack((part1, part2, part3, part4))
```

```python
n = int(input())
for i in range (n+1):
    plt.axis('equal')
    plt.plot(f(i)[0,:], f(i)[1,:])
    plt.show()
```
