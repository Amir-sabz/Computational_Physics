```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from math import floor
```

```python
small_rot = sqrt(2)/2*np.array([[1,-1],
                             [1,1]])

big_rot = sqrt(2)/2*np.array([[-1,-1],
                             [1,-1]])

def f1(point):
    return sqrt(2)/2*np.matmul(small_rot, point) + np.array([-1/2,0])
def f2(point):
    return sqrt(2)/2*np.matmul(big_rot, point) + np.array([0,-1/2])
    
func_list = [f1, f2]
```

```python
r = np.random.rand(2)
points = []
loop_count = 0
point_count = 0
for point_count in range(200000):
    for loop_count in range(11):
        i = np.random.randint(2)
        r = func_list[i](r)
        loop_count += 1
    points.append(r)
    point_count += 1
points = np.array(points)
```

```python
data = np.transpose(points)
plt.scatter(data[0], data[1], s=0.1)
plt.show()
```
