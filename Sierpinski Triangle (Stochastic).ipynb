```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def f1(point):
    return 0.5*point
def f2(point):
    return 0.5*point + np.array([1/4,sqrt(3)/4])
def f3(point):
    return 0.5*point + np.array([1/2,0])

func_list = [f1, f2, f3]
```

```python
r = np.random.rand(2)
points = []
loop_count = 0
point_count = 0
for point_count in range(100000):
    for loop_count in range(10):
        i = np.random.randint(3)
        r = func_list[i](r)
        loop_count += 1
    points.append(r)
    point_count += 1
points = np.array(points)
```

```python
data = np.transpose(points)
plt.scatter(data[0], data[1], s=0.1)
```
