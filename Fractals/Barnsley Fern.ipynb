```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sqrt
```

```python
def Transform(point, R, T):
    return np.matmul(np.array(R), point) + np.array(T)
```

```python
def f1(point):
    return Transform(point, [[0, 0],[0, 0.16]], [0,0])

def f2(point):
    return Transform(point, [[0.85, 0.04],[-0.04, 0.85]], [0, 1.6])

def f3(point):
    return Transform(point, [[0.2, -0.26],[0.23, 0.22]], [0, 1.6])

def f4(point):
    return Transform(point, [[-.15, .28],[.26, .24]], [0, .44])

func_dict = {
    1:f1,
    2:f2,
    3:f3,
    4:f4
}
```

```python
points = []
for _ in range(300000):
    r = np.random.rand(2,)
    for _ in range(10):
        i = np.random.randint(1,5)
        r = func_dict[i](r)
    points.append(r)
```

```python
points = np.array(points)
```

```python
plt.axis('equal')
plt.scatter(points[:,0], points[:,1], s=0.1, c='Green')
```
