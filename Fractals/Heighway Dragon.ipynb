```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, floor
```

```python
small_rot = sqrt(2)/2*np.array([[1,-1],
                                [1,1]])

big_rot = sqrt(2)/2*np.array([[-1,-1],
                              [1,-1]])

right_angle_rot = np.array([[0, -1],
                            [1, 0]])

def Transform(arr, a, R, T):
    return a*np.matmul(R, arr) + np.array(T)
```

```python
def f(n):
    if n==1:
        return 1/2*np.array([[-1, 0, 1],[-1, 0, -1]])
    else:
        left = Transform(f(n-1), sqrt(2)/2, small_rot, [[-1/2],[0]])
        right = Transform(f(n-1), sqrt(2)/2, big_rot, [[0],[-1/2]])[:,::-1]
        return np.hstack((left, right))
```

```python
n = int(input())
```

```python
middle = int(f(n).shape[1]/2)
plt.axis('equal')
plt.plot(f(n)[0,:middle], f(n)[1,:middle])
plt.plot(f(n)[0,middle:], f(n)[1, middle:], color='Red')
#plt.plot(f(n)[0,middle:], f(n)[1,middle:])
```
