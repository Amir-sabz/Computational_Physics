```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
rot_plus = 0.5*np.array([[1,-sqrt(3)],
                        [sqrt(3),1]])

rot_minus = 0.5*np.array([[1,sqrt(3)],
                        [-sqrt(3),1]])

Identity = np.array([[1,0],
                    [0,1]])

def rescale(point):
    return 1/3*point

def rotation_plus(point):
    return 1/3*np.matmul(rot_plus, point) + np.array([1/3,0])

def rotation_minus(point):
    return 1/3*np.matmul(rot_minus, point) + np.array([1/2,sqrt(3)/6])

def displace(point):
    return 1/3*point + np.array([2/3,0])

func_list = [rescale, rotation_plus, rotation_minus, displace]
```

```python
r = np.random.rand(2)
points = []
loop_count = 0
point_count = 0
for point_count in range(50000):
    for loop_count in range(11):
        i = np.random.randint(4)
        r = func_list[i](r)
        loop_count += 1
    points.append(r)
    point_count += 1
points = np.array(points)
```

```python
data = np.transpose(points)
plt.scatter(data[0], data[1], s=0.1)
plt.axis('equal')
plt.show()
```
